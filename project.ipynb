```python
#Sales Data Analysis

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#pandas: data ko table (DataFrame) ke form me handle karne ke liye
#numpy: calculations ke liye
#matplotlib: graphs banane ke liye
```

```python
data = pd.read_csv('sales_data.csv')

#CSV file ko read karke DataFrame banata hai
```

```python
print(data.head())

#first 5 rows show karta hai
```

```python
print(data.isnull())

#missing values check karta hai
```

```python
data = data.dropna()

#missing values wali rows remove karta hai
```

```python
data['Total_Sales'] = data['Quantity'] * data['Price']

#Quantity Ã— Price = Total sales per order
```

```python
print(data)
```

```python
region_sales = data.groupby('Region')['Total_Sales'].sum()


#groupby(): region wise data group karta hai
#sum(): total sales calculate karta hai
```

```python
#bar graph region wise sales dikhata hai
#Highest sales kaunsa region generate kar raha hai
#Kaunse products zyada sell ho rahe hain
#Business decision ke liye clear insights milte hain


region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
sales = pd.read_csv('sales_data.csv')
products = pd.read_csv('product_data.csv')

#Yahan hum dono CSV files ko alag-alag DataFrame me load kar rahe hain
```

```python
merged_data = pd.merge(sales, products, on='Product_ID', how='inner')

#pd.merge(): do DataFrame ko join karta hai
#sales, products: join hone wale tables
#on='Product_ID': common column jiske base par join ho raha hai
#how='inner': sirf matching records lega
```

```python
merged_data['Total_Sales'] = merged_data['Quantity'] * merged_data['Price']

#Har order ki total sales calculate hoti hai
```

```python
product_sales = merged_data.groupby('Product_Name')['Total_Sales'].sum()
print(product_sales)

#groupby(): product ke according data group karta hai
#sum(): total sales nikalta hai
```

```python
#Kaunsa product sabse zyada sell hua
#Kaunsa region strong performance de raha hai
#Product aur region ke basis par business decision liya ja sakta hai

product_sales.plot(kind='bar')
plt.title('Product Wise Total Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.show()
```
